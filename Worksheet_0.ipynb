Create a Python program that converts between different units of measurement.
• The program should:
1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).
2. Ask the user to input the value to be converted.
3. Perform the conversion and display the result.
4. Handle potential errors, such as invalid input or unsupported conversion types.
• Requirements:
1. Functions: Define at least one function to perform the conversion.
2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values).
3. User Input: Prompt the user to select the conversion type and input the value.
4. Docstrings: Include a docstring in your function to describe its purpose, parameters, and
return value.
• Conversion Options:
1. Length:
– Convert meters (m) to feet (ft).
– Convert feet (ft) to meters (m).
2. Weight:
– Convert kilograms (kg) to pounds (lbs).
– Convert pounds (lbs) to kilograms (kg).
3. Volume:
– Convert liters (L) to gallons (gal).
– Convert gallons (gal) to liters (L).

```python
def convert_length(value, from_unit, to_unit):
    """
    Convert length between meters and feet.

    Parameters:
    value (float): The value to be converted.
    from_unit (str): The unit of the input value ('m' or 'ft').
    to_unit (str): The unit to convert to ('m' or 'ft').

    Returns:
    float: The converted value.

    Raises:
    ValueError: If the input units are not supported.
    """
    if from_unit == 'm' and to_unit == 'ft':
        return value * 3.28084
    elif from_unit == 'ft' and to_unit == 'm':
        return value / 3.28084
    else:
        raise ValueError("Unsupported conversion type. Please use 'm' or 'ft'.")
def convert_weight(value, from_unit, to_unit):
    """
    Convert weight between kilograms and pounds.

    Parameters:
    value (float): The value to be converted.
    from_unit (str): The unit of the input value ('kg' or 'lbs').
    to_unit (str): The unit to convert to ('kg' or 'lbs').

    Returns:
    float: The converted value.

    Raises:
    ValueError: If the input units are not supported.
    """
    if from_unit == 'kg' and to_unit == 'lbs':
        return value * 2.20462
    elif from_unit == 'lbs' and to_unit == 'kg':
        return value / 2.20462
    else:
        raise ValueError("Unsupported conversion type. Please use 'kg' or 'lbs'.")

def convert_volume(value, from_unit, to_unit):
    """
    Convert volume between liters and gallons.

    Parameters:
    value (float): The value to be converted.
    from_unit (str): The unit of the input value ('L' or 'gal').
    to_unit (str): The unit to convert to ('L' or 'gal').

    Returns:
    float: The converted value.

    Raises:
    ValueError: If the input units are not supported.
    """
    if from_unit == 'L' and to_unit == 'gal':
        return value * 0.264172
    elif from_unit == 'gal' and to_unit == 'L':
        return value / 0.264172
    else:
        raise ValueError("Unsupported conversion type. Please use 'L' or 'gal'.")
def main():
    print("Welcome to the Unit Converter!")
    print("Choose the type of conversion:")
    print("1. Length (meters <-> feet)")
    print("2. Weight (kilograms <-> pounds)")
    print("3. Volume (liters <-> gallons)")

    choice = input("Enter your choice (1/2/3): ")

    try:
        if choice == '1':
            value = float(input("Enter the value to convert: "))
            from_unit = input("Convert from (m/ft): ")
            to_unit = input("Convert to (m/ft): ")
            result = convert_length(value, from_unit, to_unit)
            print(f"Converted value: {result:.2f} {to_unit}")
        elif choice == '2':
            value = float(input("Enter the value to convert: "))
            from_unit = input("Convert from (kg/lbs): ")
            to_unit = input("Convert to (kg/lbs): ")
            result = convert_weight(value, from_unit, to_unit)
            print(f"Converted value: {result:.2f} {to_unit}")
        elif choice == '3':
            value = float(input("Enter the value to convert: "))
            from_unit = input("Convert from (L/gal): ")
            to_unit = input("Convert to (L/gal): ")
            result = convert_volume(value, from_unit, to_unit)
            print(f"Converted value: {result:.2f} {to_unit}")
        else:
            print("Invalid choice. Please select 1, 2, or 3.")
    except ValueError as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```

Create a Python program that performs various mathematical operations on a list of numbers.
• The Program should:
1. Prompt the user to choose an operation (e.g., find the sum, average, maximum, or minimum
of the numbers).
2. Ask the user to input a list of numbers (separated by spaces).
3. Perform the selected operation and display the result.
4. Handle potential errors, such as invalid input or empty lists.
• Requirements:

1. Functions: Define at least one function for each operation (sum, average, maximum, mini-
mum).

2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values or
empty lists).
3. User Input: Prompt the user to select the operation and input the list of numbers.
4. Docstrings: Include a docstring in each function to describe its purpose, parameters, and
return value.

```python
def calculate_sum(numbers):
    """
    Calculate the sum of a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The sum of the numbers in the list.
    """
    return sum(numbers)
def calculate_average(numbers):
    """
    Calculate the average of a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The average of the numbers in the list.
    """
    if not numbers:
        raise ValueError("The list of numbers is empty.")
    return sum(numbers) / len(numbers)
def calculate_maximum(numbers):
    """
    Calculate the maximum value in a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The maximum value in the list.
    """
    if not numbers:
        raise ValueError("The list of numbers is empty.")
    return max(numbers)

def calculate_minimum(numbers):
    """
    Calculate the minimum value in a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The minimum value in the list.
    """
    if not numbers:
        raise ValueError("The list of numbers is empty.")
    return min(numbers)
def main():
    print("Welcome to the Mathematical Operations Program!")
    print("Choose an operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    choice = input("Enter your choice (1/2/3/4): ")

    try:
        numbers_input = input("Enter a list of numbers (separated by spaces): ")
        numbers = [float(num) for num in numbers_input.split()]

        if choice == '1':
            result = calculate_sum(numbers)
            print(f"The sum of the numbers is: {result}")
        elif choice == '2':
            result = calculate_average(numbers)
            print(f"The average of the numbers is: {result}")
        elif choice == '3':
            result = calculate_maximum(numbers)
            print(f"The maximum value is: {result}")
        elif choice == '4':
            result = calculate_minimum(numbers)
            print(f"The minimum value is: {result}")
        else:
            print("Invalid choice. Please select 1, 2, 3, or 4.")
    except ValueError as e:
        print(f"Error: {e}")

main()
```

2. Slice a Sublist:
Write a Python function that returns a sublist from a given list, starting from a specified index and
ending at another specified index.
• Requirements:
– Define a function get sublist(lst, start, end) that takes a list lst, a starting index
start, and an ending index end as input and returns the sublist from start to end (inclusive).
– Example: For the input [1, 2, 3, 4, 5, 6] with start=2 and end=4, the output should
be [3, 4, 5].

```python


def get_sublist(lst, start, end):
    """
    Return a sublist from a given list, starting from a specified index and ending at another specified index.

    Parameters:
    lst (list): The input list from which to extract the sublist.
    start (int): The starting index of the sublist (inclusive).
    end (int): The ending index of the sublist (inclusive).

    Returns:
    list: A sublist from start to end (inclusive).

    Raises:
    ValueError: If start or end indices are out of range.
    """
    if start < 0 or end >= len(lst) or start > end:
        raise ValueError("Invalid start or end index.")
    return lst[start:end + 1]
def main():
    input_list = input("Enter a list of elements (separated by spaces): ")
    lst = input_list.split()
    start = int(input("Enter the starting index: "))
    end = int(input("Enter the ending index: "))

    try:
        result = get_sublist(lst, start, end)
        print(f"Sublist from index {start} to {end}: {result}")
    except ValueError as e:
        print(f"Error: {e}")
if __name__ == "__main__":
    main()
```

4.3 Exercise on Nested List:

Write a Python function that takes a nested list and flattens it into a single list, where all the elements
are in a single dimension.
• Requirements:
– Define a function flatten(lst) that takes a nested list lst and returns a flattened version
of the list.
– Example: For the input [[1, 2], [3, 4], [5]], the output should be [1, 2, 3, 4, 5].

```python
def flatten(lst):
    """
    Flatten a nested list into a single list.

    Parameters:
    lst (list): The input nested list to be flattened.

    Returns:
    list: A new list that is a flattened version of the input list.
    """
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(flatten(item))
        else:
            flat_list.append(item)
    return flat_list
def main():
    input_list = input("Enter a nested list (e.g., [[1, 2], [3, 4], [5]]): ")
    nested_list = eval(input_list)
    result = flatten(nested_list)
    print(f"Flattened list: {result}")
if __name__ == "__main__":
    main()
```

– Example: For the input lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]] with indices = [1,

```python
def access_nested_element(lst, indices):
    """
    Extract a specific element from a nested list given its indices.

    Parameters:
    lst (list): The input nested list.
    indices (list): A list of indices that specify the position of the element to extract.

    Returns:
    The element at the specified position in the nested list.

    Raises:
    ValueError: If the indices are out of range or if the structure of the nested list does not match the indices.
    """
    try:
        element = lst
        for index in indices:
            element = element[index]
        return element
    except (IndexError, TypeError) as e:
        raise ValueError("Invalid indices or nested list structure.") from e
def main():
    input_list = input("Enter a nested list (e.g., [[1, 2], [3, 4], [5]]): ")
    nested_list = eval(input_list)
    indices_input = input("Enter the indices to access (separated by spaces): ")
    indices = [int(i) for i in indices_input.split()]
    try:
        result = access_nested_element(nested_list, indices)
        print(f"The element at the specified indices is: {result}")
    except ValueError as e:
        print(f"Error: {e}")
if __name__ == "__main__":
    main()
```

```python
def deep_flatten(lst):
    """
    Flatten a deeply nested list into a single list, regardless of the depth.

    Parameters:
    lst (list): The input deeply nested list to be flattened.

    Returns:
    list: A new list that is a flattened version of the input list.
    """
    flat_list = []
    for item in lst:
        if isinstance(item, (list, tuple)): # Updated to handle both lists and tuples
            flat_list.extend(deep_flatten(item))
        else:
            flat_list.append(item)
    return flat_list

def main():
    input_list = input("Enter a deeply nested list (e.g., [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]): ")
    nested_list = eval(input_list)
    result = deep_flatten(nested_list)
    print(f"Flattened list: {result}")

if __name__ == "__main__":
    main()
```

Complete the following tasks:
1. Create an array with values ranging from 10 to 49. {Hint:np.arrange()}.
2. Create a 3X3 matrix with values ranging from 0 to 8.
{Hint:look for np.reshape()}
3. Create a 3X3 identity matrix.{Hint:np.eye()}
4. Create a random array of size 30 and find the mean of the array.
{Hint:check for np.random.random() and array.mean() function}
5. Create a 10X10 array with random values and find the minimum and maximum values.
6. Create a zero array of size 10 and replace 5th element with 1.
7. Reverse an array arr = [1,2,0,0,4,0].
8. Create a 2d array with 1 on border and 0 inside.
9. Create a 8X8 matrix and fill it with a checkerboard pattern.

```python
import numpy as np

# 1. Create an array with values ranging from 10 to 49
array_10_to_49 = np.arange(10, 50)
print("Array with values from 10 to 49:")
print(array_10_to_49)

# 2. Create a 3X3 matrix with values ranging from 0 to 8
matrix_3x3 = np.arange(9).reshape(3, 3)
print("\n3x3 Matrix with values from 0 to 8:")
print(matrix_3x3)

# 3. Create a 3X3 identity matrix
identity_matrix = np.eye(3)
print("\n3x3 Identity Matrix:")
print(identity_matrix)

# 4. Create a random array of size 30 and find the mean of the array
random_array = np.random.random(30)
mean_value = random_array.mean()
print("\nRandom Array of size 30:")
print(random_array)
print("Mean of the random array:", mean_value)

# 5. Create a 10X10 array with random values and find the minimum and maximum values
random_10x10 = np.random.random((10, 10))
min_value = random_10x10.min()
max_value = random_10x10.max()
print("\n10x10 Array with random values:")
print(random_10x10)
print("Minimum value in the 10x10 array:", min_value)
print("Maximum value in the 10x10 array:", max_value)

# 6. Create a zero array of size 10 and replace 5th element with 1
zero_array = np.zeros(10)
zero_array[4] = 1  # Replace the 5th element (index 4) with 1
print("\nZero array of size 10 with 5th element replaced by 1:")
print(zero_array)

# 7. Reverse an array arr = [1, 2, 0,0, 4, 0]
arr = np.array([1, 2, 0, 0, 4, 0])
reversed_arr = arr[::-1]
print("\nOriginal array:", arr)
print("Reversed array:", reversed_arr)

# 8. Create a 2d array with 1 on border and 0 inside
border_array = np.ones((5, 5))
border_array[1:-1, 1:-1] = 0
print("\n2D array with 1 on border and 0 inside:")
print(border_array)

# 9. Create a 8X8 matrix and fill it with a checkerboard pattern
checkerboard = np.zeros((8, 8), dtype=int)
checkerboard[1::2, ::2] = 1  # Fill odd rows with 1s
checkerboard[::2, 1::2] = 1  # Fill even rows with 1s
print("\n8x8 Checkerboard pattern:")
print(checkerboard)
```

```python
import numpy as np
# Define the arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

# 1. Add the two arrays
add_xy = x + y
print("Addition of x and y:")
print(add_xy)

# 2. Subtract the two arrays
subtract_xy = x - y
print("\nSubtraction of x and y:")
print(subtract_xy)

# 3. Multiply the array with any integers of your choice
mult_x = x * 3  # Multiplying x by 3
print("\nMultiplication of x by 3:")
print(mult_x)

# 4. Find the square of each element of the array
square_x = x ** 2
print("\nSquare of each element in x:")
print(square_x)

# 5. Find the dot product between: v(and)w ; x(and)v ; x(and)y
dot_vw = np.dot(v, w)
print("\nDot product of v and w:")
print(dot_vw)
dot_xv = np.dot(x, v)
print("\nDot product of x and v:")
print(dot_xv)
dot_xy = np.dot(x, y)
print("\nDot product of x and y:")
print(dot_xy)

# 6. Concatenate x(and)y along row and Concatenate v(and)w along column
concat_xy_row = np.concatenate((x, y), axis=0)  # Concatenate along rows
print("\nConcatenation of x and y along rows:")
print(concat_xy_row)
concat_vw_col = np.concatenate((v.reshape(-1, 1), w.reshape(-1, 1)), axis=1)  # Concatenate along columns
print("\nConcatenation of v and w along columns:")
print(concat_vw_col)

# 7. Concatenate x(and)v; if you get an error, observe and explain why did you get the error?
try:
    concat_xv = np.concatenate((x, v), axis=0)
    print("\nConcatenation of x and v:")
    print(concat_xv)
except ValueError as e:
    print("\nError while concatenating x and v:")
    print(e)
    print("Explanation: The error occurs because x is a 2D array with shape (2, 2) and v is a 1D array with shape (2,). They cannot be concatenated along the specified axis due to their different dimensions.")
```
